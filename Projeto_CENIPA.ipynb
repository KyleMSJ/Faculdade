# *Merge* entre as tabelas

```python
from IPython.display import display
import pandas as pd

df1 = pd.read_excel("ocorrencia1_aeronauticos.xlsx")
df2 = pd.read_excel("Aeronave.xlsx")
df3 = pd.read_excel("ocorrencia_tipo.xlsx")
df4 = pd.read_excel("Fator_contribuinte.xlsx")

tb1 = pd.merge(df1, df2, how = 'inner', on = 'codigo_ocorrencia')
tb1 = pd.merge(tb1, df3, how = 'inner', on = 'codigo_ocorrencia')
tb1 = pd.merge(tb1, df4, how = 'inner', on = 'codigo_ocorrencia')
tb1.to_excel("tabelao.xlsx")
```

### Análise por localização e período

```python
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df_local = pd.DataFrame(tb1, columns=['ocorrencia_cidade', 'ocorrencia_uf'])
# Realiza a contagem com base nas colunas de cidade e estado
count_cities = df_local.groupby(['ocorrencia_cidade', 'ocorrencia_uf']).value_counts().reset_index(name='Ocorrências') # redefinindo o índice
count_cities = count_cities.rename(columns={'ocorrencia_cidade': 'Cidade','ocorrencia_uf': 'Estado'}).sort_values(by='Ocorrências', ascending=False)
count_uf = df_local['ocorrencia_uf'].value_counts().reset_index() # Retorna uma série contendo contagens de valores únicos.
count_uf = count_uf.rename(columns={'index': 'Estado', 'ocorrencia_uf': 'Ocorrências'})

# Exibe o resultado com aliases para cada coluna
print("Ocorrências por cidade:\n")
print(count_cities.head(10).to_string(index=False))
print("\n\n\nOcorrências por Estado:\n")
print(count_uf.head(10).to_string(index=False))
```

Neste código, utilizamos o método `value_counts()` para contar as ocorrências por cidade e por estado. O método `reset_index()` é utilizado para redefenir o índice e assim ordenar adequadamente conforme o número de ocorrências.

Ao imprimir o resultado, usamos o método `to_string()` para exibir o DataFrame com os cabeçalhos especificados.



```python
# Plotar o gráfico de barras
fig = px.bar(count_cities.head(10), x='Cidade', y='Ocorrências', labels={'index': 'Cidade', 'ocorrencia_cidade': 'Número de Ocorrências'},
             title='Número de Ocorrências por Cidade', width=1000, height=700)
fig.update_xaxes(tickangle=45)

fig.show()
```

### Análise por faixa de hora

```python
display(tb1.head())
```

```python
df_data = pd.DataFrame(tb1, columns=['ocorrencia_dia', 'ocorrencia_hora'])

print(df_data)
```

```python
count_time = df_data.groupby(['ocorrencia_dia', 'ocorrencia_hora']).value_counts().reset_index(name='Ocorrências')
count_time = count_time.rename(columns={'ocorrencia_dia': 'Dia','ocorrencia_hora': 'Horário'}).sort_values(by='Ocorrências', ascending=False)

print("Ocorrências em cada horário:\n")
print(count_time.head(10).to_string(index=False))
```

```python
from datetime import time
# Converter a coluna 'ocorrencia_hora' para representação numérica
df_data['Hora_Numérica'] = df_data['ocorrencia_hora'].apply(lambda t: t.hour + t.minute / 60 + t.second / 3600)

# Criar faixas de horário
faixas_horario = []
for i in range(24):
    faixas_horario.append(f"{i:02d}:00:01 - {i+1:02d}:00:00")

# Agrupar as ocorrências nas faixas de horário e calcular a contagem
df_data['Faixa_Horario'] = pd.cut(df_data['Hora_Numérica'], bins=range(24), labels=faixas_horario[:-1]) # labels faz a formatação conforme o laço de repetição acima
count_faixas = df_data.groupby('Faixa_Horario')['ocorrencia_hora'].count().reset_index(name='Ocorrências').sort_values(by='Ocorrências', ascending=False)

# Imprimir o resultado
print("Ocorrências em cada faixa de horário:\n")
print(count_faixas.head(10).to_string(index=False))
```

```python
df_data_ordered = df_data.sort_values(by='Faixa_Horario', ascending=True)
px.histogram(df_data_ordered['Faixa_Horario'])
```

```python

fig = px.bar(count_faixas.head(10), x='Faixa_Horario', y='Ocorrências', labels={'index': 'Horário', 'ocorrencia_hora': 'Número de Ocorrências'},
             title='Horários com mais ocorrências', width=1200, height=600)
fig.update_xaxes(tickangle=45)

fig.show()
```



# Análise por tipo de falha

```python
df_fator = pd.DataFrame(df4, columns=['fator_area', 'fator_nome'])
# Realiza a contagem com base nas colunas do tipo do fator e do fator em si
count_tipo_fator = df_fator['fator_area'].value_counts().reset_index()
count_tipo_fator = count_tipo_fator.rename(columns={'index': 'Tipo', 'fator_area': 'Ocorrências'})

count_fator = df_fator.groupby(['fator_area']).value_counts().reset_index(name='Ocorrências')
count_fator = count_fator.rename(columns={'fator_area': 'Tipo','fator_nome': 'Fator'}).sort_values(by=['Ocorrências'], ascending=False)



print("Ocorrências por tipo de fator:\n")
print(count_tipo_fator.head(20).to_string(index=False))

print("\n\n\n")

print("Ocorrências por fator específico:\n")
print(count_fator.head(10).to_string(index=False))


```

Realizando a análise nessa parte dos dados, foi possível verificar que a maior parte dos acidentes são causados por fatores humanos. Tanto o tipo "FATOR HUMANO" quanto "FATOR OPERACIONAL" estão diretamente relacionados com a capacidade humana, pois dentro do "FATOR HUMANO" estão aspectos como cansaço, uso de medicamentos e no "FATOR OPERACIONAL" são encontrados fatores como má utilização de instrumentos aeronáuticos ou inexperiência.

No tipo "FATOR MATERIAL" estão os aspectos de falhas mecânicas, enquanto no tipo "OUTRO" são encontrados fatores como influência do meio-ambiente.


Finalmente, é possível observar que o fator que mais causou acidentes dentro do banco de dados verificado foi "JULGAMENTO DE PILOTAGEM".

```python
fig = px.bar(count_tipo_fator.head(10), x='Tipo', y='Ocorrências', labels={'index': 'Tipo', 'fator_area': 'Número de Ocorrências'},
             title='Número de Ocorrências por tipo', width=800, height=750)
fig.show()
```

```python
fig = px.bar(count_fator.head(10), x='Fator', y='Ocorrências', labels={'index': 'Fator', 'fator_nome': 'Número de Ocorrências'},
             title='Número de Ocorrências por Fator Específico', width=1000, height=700)
fig.update_xaxes(tickangle=45)

fig.show()
```
